```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame ( {'x_1':['A','B','C']*3,
                      'y_1':['D','E','F']*3,
                      'z':np.random.randint(0,20,9)})
df1
```

```python
df2 = pd.DataFrame ( {'x_2':['A','B','C']*4,
                      'y_2':['D','E',]*6,
                      'z':np.random.randint(0,20,12)})
df2
```

```python
df3 = pd.merge (df1, df2,how ='inner',left_on = ['x_1','y_1'],right_on = ['x_2','y_2'],suffixes=['_left','_right'])
```

```python
df3
```

```python
df4 = pd.merge(df1, df2, 
                how='outer',
                left_on=['x_1', 'y_1'],
                right_on=['x_2', 'y_2'],
                suffixes=['_left', '_right'])
```

```python
df4
```

```python
df4.fillna(0,inplace=True)
```

```python
z_avg = (df4.z_left.add(df4.z_right,fill_value=0))/2

df4_final = df4.join(z_avg.to_frame('z_avg'))
df4_final
```
