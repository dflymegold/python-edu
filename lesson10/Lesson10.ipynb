```python
import numpy as np
import scipy.integrate
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

class IntegralCalculus :
  def __init__ (self,function , left_limit, right_limit):
    self.function = function
    self.left_limit = left_limit
    self.right_limit = right_limit

  def describeFigure (self):
    left = print('left_limit is = ',self.left_limit) 
    right = print('right limit is = ',self.right_limit)
    function = print ('function is = ',self.function)  
    return 0
  def integrateFigure (self):
    if self.function == 'sin(x)':
      x2 = lambda x : np.sin(x)
    elif self.function == 'sin(x)+cos(2x)' :
        x2 = lambda x: np.sin(x)+np.cos(2*x)
    elif self.function == 'x*tg(x)+2' :
        x2 = lambda x: x*np.tan(x)+2
    elif self.function == 'x^3-10*x^2+170' :
        x2 = lambda x: x**3-10*x**2+170
    rez =scipy.integrate.quad(x2,self.right_limit,self.left_limit)
    return rez
  def plotFigure(self):
    if self.function == 'sin(x)':
      x2 = lambda x : np.sin(x)
    elif self.function == 'sin(x)+cos(2x)' :
        x2 = lambda x: np.sin(x)+np.cos(2*x)
    elif self.function == 'x*tg(x)+2' :
        x2 = lambda x: x*np.tan(x)+2
    elif self.function == 'x^3-10*x^2+170' :
        x2 = lambda x: x**3-10*x**2+170
    x = np.linspace (0,10)
    y = x2(x)
    fig, ax = plt.subplots()
    ax.plot(x, y, 'r', linewidth=2)
    ax.set_ylim(bottom=0)
    ix = np.linspace(self.left_limit, self.right_limit)
    iy = x2(ix)
    verts = [(self.left_limit, 0), *zip(ix, iy), (self.right_limit, 0)]
    poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
    ax.add_patch(poly)
    ax.text(0.5 * (self.left_limit + self.right_limit), 30, r"$\int_a^b f(x)\mathrm{d}x$",
        horizontalalignment='center', fontsize=20)

    fig.text(0.9, 0.05, '$x$')
    fig.text(0.1, 0.9, '$y$')

    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)
    ax.xaxis.set_ticks_position('bottom')

    ax.set_xticks([self.left_limit, self.right_limit], labels=['$a$', '$b$'])
    ax.set_yticks([])
    return 0

if __name__ == "__main__":
  function_1 = IntegralCalculus('sin(x)',0,1.2)
  plotter1 = function_1.describeFigure()
  findplot1 = function_1.integrateFigure()
  print (findplot1[0])
  function_2 = IntegralCalculus('sin(x)+cos(2x)',0,1.2)
  plotter2 = function_2.describeFigure()
  findplot2 = function_2.integrateFigure()
  print (findplot2[0])
  function_3 = IntegralCalculus('x*tg(x)+2',0.2,0.8)
  plotter3 = function_3.describeFigure()
  findplot3 = function_3.integrateFigure()
  print (findplot3[0])
  function_4 = IntegralCalculus('x^3-10*x^2+170',0,10)
  plotter4 = function_4.describeFigure()
  findplot4 = function_4.integrateFigure()
  print (findplot4[0])
  plot = function_1.plotFigure()
  plt.show()
```
