```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("/content/dataset.txt") as file:
  data = [row.strip() for row in file]
  data =  np.delete(data,0)
  x = []
  y = []
  for _ in data:
        x.append(_.split(" ")[0])
  for _ in data:
        y.append(_.split(" ")[1])
x_1 = list(map(int, x))
y_1 = list (map(float,y))
```

```python
plt.plot(x_1,y_1)
plt.title('Linear graph', fontsize=15)
plt.xlabel('x', fontsize=15, color='black')
plt.ylabel('y', fontsize=15, color='black')
plt.grid(True)
plt.show()
```

```python
where_set = ['pre', 'post', 'mid']
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
for i, ax in enumerate(axs):
  ax.step(x_1, y_1, 'g-o', where=where_set[i])
  ax.grid(True)
```

```python
plt.stem(x_1, y_1)
plt.title('Steam graph', fontsize=15)
plt.xlabel('x', fontsize=15, color='black')
plt.ylabel('y', fontsize=15, color='black')
plt.grid(True)
plt.show()
```

```python
plt.barh(x_1, y_1)
plt.title('Bar graph', fontsize=15)
plt.xlabel('x', fontsize=15, color='black')
plt.ylabel('y', fontsize=15, color='black')
plt.grid()
plt.show()
```

```python
y_max = max(y_1)
print (y_max)
index =y_1.index(y_max)
x_max = x_1[index]
print (x_max)
plt.annotate(("max",x_max,y_max),xy = (x_max,y_max))
plt.plot (x_1,y_1)
plt.grid()
plt.show()
```

```python
y_2 = [i * 2 for i in y_1]

plt.plot(x_1, y_1, '-')
plt.plot(x_1, y_2, '--')
plt.title('Linear graph of y(x) and 2*y(x)', fontsize=15)
plt.xlabel('x', fontsize=15, color='black')
plt.ylabel('y', fontsize=15, color='black')
plt.grid(True)
plt.show()
```
