```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = pd.read_csv("/content/dataset.csv")

df1 = pd.DataFrame(data)
```

```python
df1.head(10)
```

```python
df1.tail(10)
```

```python
y = df1.iloc[:,-1]
```

```python
y
```

```python
X = df1.iloc[:,[i for i in range(20)]]
```

```python
X
```

```python
le = preprocessing.LabelEncoder()
le.fit(["male","female"])
y_decoded = le.transform(y)
```

```python
scaler = preprocessing.StandardScaler()
scaler.fit(X)
X_standarted = scaler.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_standarted,y_decoded,test_size=0.2,train_size = 0.8)
```

```python
clf = make_pipeline( SVC(kernel = 'linear'))
clf.fit(X_train, y_train)
```

```python
predicted = clf.predict(X_test)
expected = y_test
```

```python
accuracy_score(predicted,expected)
```

```python
clfd = make_pipeline( SVC(kernel = 'rbf'))
clfd.fit(X_train, y_train)
```

```python
predictedf = clfd.predict(X_test)
expectedf = y_test
```

```python
accuracy_score(predictedf,expectedf)
```
